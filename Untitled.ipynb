```python
exp.set_tags({"tag1":"classification"})
exp.set_tags({"tag2":"Breast Cancer"})
exp.set_tags({"tag3":"PCA"})
```

```python
def train_model(X, y, run):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    
    # Make an instance of the Model
    pca = PCA()
    
    pca.fit(X_train)
    X_train = pca.transform(X_train)
    X_test = pca.transform(X_test)
    
    model = LogisticRegression()
    print('training')
    start_time = time.time()
    model.fit(X_train, y_train)
    duration = time.time() - start_time
    print('done')
    run.log("Number of component", component_number=100)
    run.log("Run time Duration", duration)
    run.log("Accuracy", model.score(X_test,y_test))
    run.complete()
    
    return model
```
